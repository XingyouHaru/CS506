f) Complete the code below to replace all missing ages with the average age

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.head()
```

g) Plot a bar chart of the average age of those that survived and did not survive. Briefly comment on what you observe.

```python
SurvivedAge = df[df['Survived'] == 1]['Age'].mean()
NotSurvivedAge = df[df['Survived'] == 0]['Age'].mean()
avg = pd.DataFrame({'whether_survive': ['Survived', 'NotSurvived'], 'age':[SurvivedAge, NotSurvivedAge]})
avg.plot(x='whether_survive', y='age', kind='bar')

# People who survived from train accident are generally younger than those who did not.
```
